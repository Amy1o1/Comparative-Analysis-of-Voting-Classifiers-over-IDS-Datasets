```python
# droping nan and inf rows
# first replacing all inf values with nan
df.replace([np.inf, -np.inf], np.nan, inplace= True)
df.info()
```

```python
df.dropna(inplace = True) # dropping all nan rows
```

```python
df.hist(bins = 50, figsize=(20, 15))
plt.show()
```

```python
import seaborn as sns

df = df.dropna('columns')# drop columns with NaN
df = df[[col for col in df if df[col].nunique() > 1]]# keep columns where there are more than 1 unique values
corr = df.corr()

plt.figure(figsize=(15,12))
sns.heatmap(corr)
plt.show()
```

```python
# 1 -----> Benign
# 0 -----> Attack

Class = []
for i in df['Class'].values:
    if i=='Benign.':
        Class.append(1)
    else:
        Class.append(0)
print(len(Class))
print(Class[:10])
```

```python
df['Class'] = Class
df.head()
```

```python
df.drop('Label', axis = 1, inplace= True)
```

```python
df.head()
```

```python
drops1 = []
drops2 = []
vals = []
for i in df.columns:
    for j in df.columns:
        if i == j:
            continue
        corr_val = df[i].corr(df[j])
        if corr_val > 0.9799999999999999:
            vals.append(corr_val)
            print(corr_val)
            if i not in drops1:
                drops1.append(i)
            if j not in drops2: 
                drops2.append(j)
```

```python
drops1
```

```python
for i in drops1:
    if 'Class' in drops1:
        drops1.remove('Class')
```

```python
df.drop(drops1, axis = 1, inplace=True)
```
